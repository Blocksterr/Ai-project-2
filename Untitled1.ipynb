```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sample_file = "sample_submission (1).csv"
train_file = "train.csv"
test_file = "test.csv"
```

```python
train_data = pd.read_csv(f'{train_file}')
test_data = pd.read_csv(f'{test_file}')
```

```python
categorical_features = ["fullAddress","postcode", "country", "outcode", "bedrooms", "bathrooms", "livingRooms", "tenure", "propertyType", "currentEnergyRating", "sale_month"]
train_data["price"] = np.log(train_data["price"])
for col in categorical_features:
    train_data[col] = train_data[col].astype("category")
    test_data[col] = test_data[col].astype("category")
```

```python
x_train, y_train = train_data.drop(columns = ["price"]), train_data["price"]
```
